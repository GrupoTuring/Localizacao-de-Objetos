```python
import Augmentation as ag
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import sys
import numpy as np
import pandas as pd
```

```python
pikachu_path = 'Test_pics/pikachu.png'
eevee_path = 'Test_pics/eevee.png'
lucario_path = 'Test_pics/lucario.png'
background_path = 'Test_pics/background.png'

pikachu_im = Image.open(pikachu_path)
eevee_im = Image.open(eevee_path)
lucario_im = Image.open(lucario_path)
background = Image.open(background_path)

fig, ax = plt.subplots(ncols=4, figsize=(15,5))
ax[0].imshow(pikachu_im)
ax[0].set_title("Pikachu")
ax[1].imshow(eevee_im)
ax[1].set_title("Eevee")
ax[2].imshow(lucario_im)
ax[2].set_title("Lucario")
ax[3].imshow(background)
ax[3].set_title("Background")

for a in ax.ravel():
    a.axis('off')
```

```python
labels = ['pikachu']
images = [pikachu_im]
ag.blendImages(objImages = images, imgLabels = labels, envImage = background, num_images = 1, path = 'results')
```
